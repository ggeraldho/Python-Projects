```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np

df = pd.read_csv('Credit card default I.csv')

df.head()
```

```python
pd.set_option('display.max_columns', None)
df = df.drop(columns = 'ID')
df.describe(include = 'all')
```

```python
#checking to see the unique values in each column
for col in df.columns:
    unique_values = df[col].unique()
    print('Unique values for {}: {}'.format(col, unique_values))
```

===================================================================================================================

In the data dictionary, we see that for PAY_0 to PAY_6, the scale runs from -1, then 1 to 9. We notice there are the values -2 and 0, which are not inside the data dictionary. So does this mean they are erroneous values? 

In this analysis, I make the assumption that -2 = 'No consumption', and 0 = 'Use of revolving credit (paying minimum credit amount only). This is due to both values taking up more than half of the dataset, which would make imputation and removal near impossible. Hence, we make these two assumptions for the values -2 and 0 in PAY_0 to PAY_6, and keep them as they are. However, education and marriage have erroneous values of 5,6,0, and 0, which we then proceed to impute

===================================================================================================================

```python
#we see from the above that education and marriage have 5,6,0 and 0 respectively. hence need to impute
df['EDUCATION'] = df['EDUCATION'].replace([0,5,6], np.nan)
df['MARRIAGE'] = df['MARRIAGE'].replace([0], np.nan)

#impute missing values with mode
columns_to_impute = ['EDUCATION', 'MARRIAGE']

# Impute the missing values with mode for the selected columns
for column in columns_to_impute:
    mode_value = df[column].mode()[0]
    df[column].fillna(mode_value, inplace=True)
```

```python
#CREATING DUMMY VARIABLES
dummy = pd.get_dummies(df["SEX"], prefix = 'SEX') #optional, set prefix=xxx
df = df.merge(dummy, right_index=True, left_index=True)
df = df.drop(columns="SEX")
dummy = pd.get_dummies(df["EDUCATION"], prefix = 'EDUCATION') #optional, set prefix=xxx
df = df.merge(dummy, right_index=True, left_index=True)
df = df.drop(columns='EDUCATION')
dummy = pd.get_dummies(df["MARRIAGE"], prefix = 'MARRIAGE') #optional, set prefix=xxx
df = df.merge(dummy, right_index=True, left_index=True)
df = df.drop(columns='MARRIAGE')
dummy = pd.get_dummies(df["PAY_0"], prefix = 'PAY0') #optional, set prefix=xxx
df = df.merge(dummy, right_index=True, left_index=True)
df = df.drop(columns='PAY_0')
dummy = pd.get_dummies(df["PAY_2"], prefix='PAY2') #optional, set prefix=xxx
df = df.merge(dummy, right_index=True, left_index=True)
df = df.drop(columns='PAY_2')
dummy = pd.get_dummies(df["PAY_3"], prefix='PAY3') #optional, set prefix=xxx
df = df.merge(dummy, right_index=True, left_index=True)
df = df.drop(columns='PAY_3')
dummy = pd.get_dummies(df["PAY_4"], prefix='PAY4') #optional, set prefix=xxx
df = df.merge(dummy, right_index=True, left_index=True)
df = df.drop(columns='PAY_4')
dummy = pd.get_dummies(df["PAY_5"], prefix='PAY5') #optional, set prefix=xxx
df = df.merge(dummy, right_index=True, left_index=True)
df = df.drop(columns='PAY_5')
dummy = pd.get_dummies(df["PAY_6"], prefix='PAY6') #optional, set prefix=xxx
df = df.merge(dummy, right_index=True, left_index=True)
df = df.drop(columns='PAY_6')

df
```

```python
#remove outliers
from sklearn.model_selection import train_test_split 
df_train,df_test = train_test_split(df)
from scipy import stats
z = np.abs(stats.zscore(df))
df = df[(z < 5).all(axis=1)]
```

```python
#checking if imbalanced
df["default payment next month"].value_counts()
```

```python
#using SMOTE to balance
Y = df["default payment next month"]
X = df.drop(columns="default payment next month")

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y)

from imblearn.over_sampling import SMOTE

Y_train.value_counts()
X_train, Y_train =SMOTE().fit_resample(X_train, Y_train)
```

```python
#normalisation
X_train['LIMIT_BAL'] = stats.zscore(X_train['LIMIT_BAL'])
X_test['LIMIT_BAL'] = stats.zscore(X_test['LIMIT_BAL'])

cols_to_normalize = df.columns[2:14]

from scipy.stats import zscore

X_train[cols_to_normalize] = X_train[cols_to_normalize].apply(zscore)

X_test[cols_to_normalize] = X_test[cols_to_normalize].apply(zscore)
```

```python
X_train #to check if normalisation successful
```

```python
#logistic regression implementation
from sklearn.metrics import confusion_matrix, roc_curve, auc, precision_recall_curve

from sklearn import linear_model
model = linear_model.LogisticRegression()
model.fit(X_train,Y_train)
pred = model.predict(X_test)
cm = confusion_matrix(Y_test,pred)
print("accuracy", (cm[0,0]+cm[1,1])/(sum(sum(cm))))
```

```python
fpr, tpr, thresholds = roc_curve(Y_test, pred)
roc_auc = auc(fpr, tpr)
precision, recall, _ = precision_recall_curve(Y_test, pred)
pr_auc = auc(recall, precision)

#to plot graphs in one row
fig, ax = plt.subplots(ncols=3, figsize=(15, 5))

#confusion matrix
im = ax[0].imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
ax[0].set_title('Confusion matrix')
ax[0].set_xlabel('Predicted label')
ax[0].set_ylabel('True label')
ax[0].set_xticks([0,1])
ax[0].set_yticks([0,1])
ax[0].set_xticklabels(['Negative', 'Positive'])
ax[0].set_yticklabels(['Negative', 'Positive'])
cbar = ax[0].figure.colorbar(im, ax=ax[0])
cbar.ax.set_ylabel('Counts', rotation=-90, va="bottom")

#ROC Curve
ax[1].plot(fpr, tpr, lw=1, label='ROC (area = %0.2f)' % roc_auc)
ax[1].plot([0, 1], [0, 1], '--', color='gray', lw=1)
ax[1].set_xlim([0, 1])
ax[1].set_ylim([0, 1])
ax[1].set_xlabel('False Positive Rate')
ax[1].set_ylabel('True Positive Rate')
ax[1].set_title('ROC')
ax[1].legend(loc="lower right")

#PRC curve
ax[2].plot(recall, precision, lw=1, label='PRC (area = %0.2f)' % pr_auc)
ax[2].set_xlim([0, 1])
ax[2].set_ylim([0, 1])
ax[2].set_xlabel('Recall')
ax[2].set_ylabel('Precision')
ax[2].set_title('PRC')
ax[2].legend(loc="lower right")

plt.show()
```

How do we improve the results of the models?

- Feature Selection: The goal of feature selection is to identify and select the most relevant features for your model. This is important because irrelevant or redundant features can negatively impact the performance of your model. Correlation-based feature selection, recursive feature elimination, and principal component analysis are some of the techniques you can use to identify and select the most relevant features. By reducing the number of features, you can reduce the noise in the data and improve the accuracy of your model.

- Data Wrangling: Data cleaning and preprocessing is an important step in building any machine learning model. Outlier detection and removal, data normalization, and feature scaling are some of the techniques you can use to improve the quality of your data. Outliers can negatively impact the performance of your model by skewing the distribution of your data. Data normalization and feature scaling can help to make the data more consistent and reduce the variance in the data.

- Parameter Setting: Each machine learning model has several hyperparameters that can be tuned to improve the performance of the model. For example, in logistic regression, you can adjust the regularization strength to prevent overfitting. In random forest, you can adjust the number of trees and the depth of the tree to control the bias and variance of the model. By trying different hyperparameters, you can tune the model for better performance and prevent overfitting or underfitting.

- Ensemble Learning: Ensemble learning is a technique where you combine the outputs of multiple models to improve the performance of the model. Bagging, boosting, and stacking are some of the techniques you can use for ensemble learning. By combining the outputs of multiple models, you can reduce the variance and bias of the model and improve the robustness and accuracy of the model.

===================================================================================================================


How do we overcome the weaknesses of the models?

Regression:
- Collinearity: One strategy to overcome the issue of collinearity is to use regularisation techniques like Lasso regression. Such techniques can reduce the impact of highly correlated features by shrinking the coefficients associated with these features.
- Limited ability to model complex relationships: One strategy to overcome this limitation is to use more complex models such as nonlinear regression. These models can capture more complex relationships between the variables.

Decision Tree:
- Overfitting: One strategy to overcome the issue of overfitting in decision trees is to prune the tree by setting a maximum depth or by setting a minimum number of samples per leaf.
- Instability: One strategy to overcome instability in decision trees is to use ensemble methods such as Random Forest or Gradient Boosted Trees. These methods can combine multiple decision trees to improve stability and reduce variance. We already see this where our random forest did better than decision tree in terms of accuracy
- Poor performance with many features: One strategy to overcome poor performance with many features is to use feature selection techniques. These techniques can reduce the dimensionality of the data and identify the most important features.

XGBoost:
- Computational Expense: One strategy to overcome the issue of computational expense in XGBoost is to use a subset of the data or features for training. This can reduce the computational requirements and speed up training time.

SKlearn and Keras/Tensorflow Neural Network:
- Overfitting: One strategy to overcome overfitting in neural networks is to use regularisation techniques such as dropout. These techniques can prevent the network from becoming too complex and overfitting the training data.
- Local Optima: One strategy to overcome the issue of local optima in neural networks is to use multiple initialisations or to use a different optimizer. These strategies can help the network to explore a wider range of solutions and avoid getting stuck in a local optima.
- Limited Data: One strategy to overcome the issue of limited data is to use data augmentation techniques such as flipping, rotating, or scaling the images. This can increase the amount of data available for training and improve the performance of the network.
